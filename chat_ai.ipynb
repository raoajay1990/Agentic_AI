```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
ollama_api_key = os.getenv("OLLAMA_API_KEY")
url = "https://api.openai.com/v1"
client = OpenAI(api_key=ollama_api_key,base_url=url)
model = "gpt-oss:120b"


```

```python
def llm_caller(prompt):
    try:
        response = client.chat.completions.create(
            model = model,
            messages = [
                {
                    "role":"system",
                    "content":"Act as export weather predictor, someone who knows and has experience on changing weather"
                },
                {
                    "role":"user",
                    "content": prompt
                }
            ]
        )
        print("\nResponse:\n", response.choices[0].message.content)
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
prompt = input("Ask your question or enter quit to stop")
while (prompt.lower() != 'quit' ):
    llm_caller(prompt)
    
```
