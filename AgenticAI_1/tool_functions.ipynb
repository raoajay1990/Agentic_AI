```python
import requests
import os
from dotenv import load_dotenv
import json
```

```python

load_dotenv()
rapid_api_key = os.getenv('RAPID_API_KEY')
url = "https://cricket-api-free-data.p.rapidapi.com/cricket-schedule"
```

```python
headers = {
    "x-rapidapi-key": rapid_api_key,
	"x-rapidapi-host": "cricket-api-free-data.p.rapidapi.com",
	"Content-Type": "application/json"
}
```

```python
tool_functions = {}
```

```python
def tool_function_decorator(name):
    def wrapper(func):
        tool_functions[name] = func
        return func
    return wrapper
```

```python
@tool_function_decorator('get_fixtures')
def get_fixtures():
    response = requests.get(url, headers = headers)
    print(response.json())

    
```

Working on Tool Registry today


```python
tool_registry = {}
tool_json_list = []
```

```python
json_directory = os.getcwd()
files = os.listdir(json_directory)

for file in files:
    if file.endswith('.json'):
        file_path = os.path.join(json_directory,file)
        with open(file_path, 'r') as f:
            tool_data = json.load(f)
            tool_json_list.append(tool_data)
            tool_name = tool_data['function']['name']
            if tool_name in tool_functions.keys():
                tool_registry[tool_name] = tool_functions[tool_name]
            else:
                print(f'tool function does not exist for this tool_name : {tool_name}')


# tool_registry
```

```python
class ToolRegistry:

    def __init__(self):
        pass

    def get_tool_registry(self):
        return tool_registry
    
    

        
```
